```python
import pandas as pd
df = pd.read_excel('Talent_Academy_Case_DT_2025.xlsx')
print(df.head())
```

```python
print("Veri Seti Bilgileri:")
print(df.info())
print("\n----------------------\n")
print("Sayısal Sütunların İstatistiksel Özeti:")
print(df.describe())
```

```python
import numpy as np

# TedaviSuresi sütununu temizleme / Seans metnini silip sayısal değerlere dönüştürme
df['TedaviSuresi'] = df['TedaviSuresi'].str.replace(' Seans', '').astype(np.int64)

# UygulamaSuresi sütununu temizleme / Dakika metnini silip sayısal değerlere dönüştürme
df['UygulamaSuresi'] = df['UygulamaSuresi'].str.replace(' Dakika', '').astype(np.int64)

# Değişiklikleri doğrulama
print("Güncellenmiş Veri Seti Bilgileri:")
print(df.info())

print("\nGüncellenmiş Veri Seti İlk 5 Satırı:")
print(df.head())
```

```python
# Eksik değerleri doldurmak için doldurulacak sütunları seçip Bilinmiyor etiketi ile doldurma
columns_with_missing_values = ['Cinsiyet', 'KanGrubu', 'KronikHastalik', 'Bolum', 'Alerji', 'Tanilar', 'UygulamaYerleri']
df[columns_with_missing_values] = df[columns_with_missing_values].fillna('Bilinmiyor')

print("Eksik değerler doldurulduktan sonraki veri seti bilgileri:")
print(df.info())
```

```python
# 'KronikHastalik' sütununu virgülle ayırıp her birini ayrı bir satıra dönüştürme
chronic_diseases = df['KronikHastalik'].str.split(', ').explode()

# Her bir kronik hastalık için dummy (0-1) değişkenler oluşturma
chronic_diseases_dummies = pd.get_dummies(chronic_diseases, prefix='Chronic')

# Orijinal dataframe ile dummy değişkenleri birleştirme
df = pd.concat([df, chronic_diseases_dummies], axis=1)

# Orijinal 'KronikHastalik' sütununu kaldırma
df = df.drop('KronikHastalik', axis=1)

print("Güncellenmiş veri setinin ilk 5 satırı:")
print(df.head())
```

```python
import pandas as pd

df = pd.read_excel('Talent_Academy_Case_DT_2025.xlsx')

# 'TedaviSuresi' ve 'UygulamaSuresi' sütunlarını yeniden temizle
df['TedaviSuresi'] = df['TedaviSuresi'].str.replace(' Seans', '').astype(int)
df['UygulamaSuresi'] = df['UygulamaSuresi'].str.replace(' Dakika', '').astype(int)

# Eksik değerleri yeniden 'Bilinmiyor' olarak doldurma
columns_with_missing_values = ['Cinsiyet', 'KanGrubu', 'KronikHastalik', 'Bolum', 'Alerji', 'Tanilar', 'UygulamaYerleri']
df[columns_with_missing_values] = df[columns_with_missing_values].fillna('Bilinmiyor')

# Sütunları doğru şekilde işleme zamanı
# 'KronikHastalik' sütunundaki her bir hastalığı için yeni sütunlar oluşturma
chronic_diseases_dummies = df['KronikHastalik'].str.get_dummies(sep=', ').add_prefix('Chronic_')
df = pd.concat([df, chronic_diseases_dummies], axis=1)
df = df.drop('KronikHastalik', axis=1)

# 'Alerji' sütunundaki her bir alerji için yeni sütunlar oluşturma
allergy_dummies = df['Alerji'].str.get_dummies(sep=',').add_prefix('Alerji_')
df = pd.concat([df, allergy_dummies], axis=1)
df = df.drop('Alerji', axis=1)

# 'Tanilar' sütunundaki her bir tanı için yeni sütunlar oluşturma
diagnosis_dummies = df['Tanilar'].str.get_dummies(sep=',').add_prefix('Tani_')
df = pd.concat([df, diagnosis_dummies], axis=1)
df = df.drop('Tanilar', axis=1)

# 'UygulamaYerleri' sütunundaki her bir uygulama yeri için yeni sütunlar oluşturma
application_site_dummies = df['UygulamaYerleri'].str.get_dummies(sep=',').add_prefix('Uygulama_')
df = pd.concat([df, application_site_dummies], axis=1)
df = df.drop('UygulamaYerleri', axis=1)

# Veri setinin yeni halini görüntüleme
print("Çoklu değer içeren sütunlar işlendikten sonraki veri seti bilgileri:")
print(df.info())

print("\nÇoklu değer içeren sütunlar işlendikten sonraki veri seti ilk 5 satırı:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(data=df, x='Yas', kde=True, ax=axes[0])
axes[0].set_title('Yaş Dağılımı', fontsize=15)
axes[0].set_xlabel('Yaş', fontsize=12)
axes[0].set_ylabel('Frekans', fontsize=12)

sns.histplot(data=df, x='TedaviSuresi', kde=True, ax=axes[1])
axes[1].set_title('Tedavi Süresi Dağılımı (Seans)', fontsize=15)
axes[1].set_xlabel('Tedavi Süresi (Seans)', fontsize=12)
axes[1].set_ylabel('Frekans', fontsize=12)

sns.histplot(data=df, x='UygulamaSuresi', kde=True, ax=axes[2])
axes[2].set_title('Uygulama Süresi Dağılımı (Dakika)', fontsize=15)
axes[2].set_xlabel('Uygulama Süresi (Dakika)', fontsize=12)
axes[2].set_ylabel('Frekans', fontsize=12)

plt.tight_layout()
plt.savefig('numerical_distributions.png')

print("Sayısal sütunların dağılımını gösteren histogramlar oluşturuldu ve 'numerical_distributions.png' olarak kaydedildi.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.boxplot(x='Cinsiyet', y='TedaviSuresi', data=df, ax=axes[0])
axes[0].set_title('Cinsiyet ve Tedavi Süresi İlişkisi', fontsize=15)
axes[0].set_xlabel('Cinsiyet', fontsize=12)
axes[0].set_ylabel('Tedavi Süresi (Seans)', fontsize=12)

sns.boxplot(x='KanGrubu', y='TedaviSuresi', data=df, ax=axes[1])
axes[1].set_title('Kan Grubu ve Tedavi Süresi İlişkisi', fontsize=15)
axes[1].set_xlabel('Kan Grubu', fontsize=12)
axes[1].set_ylabel('Tedavi Süresi (Seans)', fontsize=12)

plt.tight_layout()
plt.savefig('categorical_relationships.png')

print("Kategorik sütunların Tedavi Süresi ile ilişkisini gösteren box plot grafikleri oluşturuldu ve 'categorical_relationships.png' olarak kaydedildi.")
```
